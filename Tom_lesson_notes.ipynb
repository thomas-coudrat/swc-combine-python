### Partial statistics on array

```python
# How about getting the max value from patient 0
patient_0 = data[0, :] # 0 on the first axis, everything on the second
print 'maximum inflammation for patient 0:', patient_0.max()
```

```python
# The row doesn't need to be stored in a variable, we can directly do:
print 'maximum inflammation for patient 2:', data[2, :].max()
```

```python
# Maximum inflammation for all patients, or the average for each day
# across axis 0: top-down
# across axis 1: left-right

# Example:
print data.mean(axis=0)
```

### Plotting

```python
# Let's import another library: matplotlib is the default plotting library for Python
from matplotlib import pyplot
# And use an IPython notebook magic command: embeds the plots in the notebook instead of in their own window
%matplotlib inline

# Create a display a heatmap of our data
image  = pyplot.imshow(data)
pyplot.show(image)

# Blue = low values, Red = high values (rise and fall of inflammation over the 40 day period)
```

```python
# Average inflammation over axis 0 (over time)
ave_inflammation = data.mean(axis=0)
ave_plot = pyplot.plot(ave_inflammation)
pyplot.show(ave_plot)
# Looks suspiciously like a linear rise & fall (we expected sharper rise, slower fall)
```

```python
# Max inflammation over the axis 0 (over time)
max_plot = pyplot.plot(data.max(axis=0))
pyplot.show(max_plot)
# Again, very linear
```

```python
# Min inflammation over the axis 0 (over time)
min_plot = pyplot.plot(data.min(axis=0))
pyplot.show(min_plot)
```

```python
# Linear rise and falls of Max and Min: problem with our analysis method or the data?
```

### Library alias, figure with subplot

```python
# Convention when importing libraries: use aliases np, plt

import numpy as np
from matplotlib import pyplot as plt

data = np.loadtxt(fname='data/inflammation-01.csv', delimiter=',')

# How large should the figure be?
fig = plt.figure(figsize=(10.0, 3.0))

# Creating three subplots
axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

# What to draw for each one
axes1.set_ylabel('average')
axes1.plot(data.mean(axis=0))

axes2.set_ylabel('max')
axes2.plot(data.max(axis=0))

axes3.set_ylabel('min')
axes3.plot(data.min(axis=0))

# We want a tight layout
# fig.tight_layout()

plt.show(fig)
```

### Strings = list of chars

```python
# Strings are actually lists of characters
# To print each character of the word "lead", we can do:
word = "lead"
print word[0]
print word[1]
print word[2]
print word[3]
```

```python
# BAD: doesn't scale (what if the word is 100 chars), AND fragile (doesn't apply to other lenght words)
word = "tin"
print word[0]
print word[1]
print word[2]
print word[3]
```

### Loop: indentation

```python
# Better approach: loop
word = 'lead'
for char in word:
    print char
# Shorter than printing each individual chars
# More robust: same code applies to any word length
```

```python
# General description:
# for loop repeats an operation (print) for each thing in a collection (each char in string)

# for variable in collection:
#     do things with variable

# The loop variable can be called anything you like
```

### Counting with a loop

```python
# Another example with vowels
length = 0
for vowel in 'aeiou':
    length = length + 1
print 'There are', length, 'vowels'

# Loop will execute operation 5 times (5 vowels): first round length = 1 ... result length = 5
```

```python
# Loop variable still exists after loop is over
length = 0
for vowel in 'aeiou':
    length = length + 1
print 'There are', length, 'vowels'
```

```python
# Length of a list: very common operation in Python
len('aeiou')
```

### 03-Lists

## Goals
* Explain what a list is.
* Create and index lists of simple values.

```python
# A for loop: do an operation many times
# List: store many values

# Built into Python, unlike NumPy arrays
```

### Deal with lists

```python
# Example
odds = [1, 3, 5, 7]
print 'odds are:', odds
```

```python
# Select first and last, using indexes
# Negative index counts from the end!
print 'first and last:', odds[0], odds[-1]
```

```python
# Loop over that list
for number in odds:
    print number
```

### Difference between lists and strings (mutable vs immutable)

```python
# typo in Darwin's name
names = ['Newton', 'Darwing', 'Turing']
print 'names is originally:', names

# change only the name 'Darwin'
names[1] = 'Darwin' # correct the name
print 'final value of names:', names
```

```python
# But changing a char in a string doesn't work
name = 'Bell'
name[0] = 'b'
```

```python
# Mutable: can be modified after being created
# Immutable: cannot be modified after being created: need to replace it with updated value
```

### List methods, function calls

```python
# Add element to the end of the list

odds.append(11)
print 'odds after adding a value:', odds
```

```python
# Delete element, at index 0

del odds[0]
print 'odds after removing the first element:', odds
```

```python
# Reverse order of the list

odds.reverse()
print 'odds after reversing:', odds
```
