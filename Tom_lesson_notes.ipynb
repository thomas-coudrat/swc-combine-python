```python
# Built-in function
dir(numpy)
```

```python
# Using a numpy function: loadtxt
# parameters: character strings in quotes
numpy.loadtxt(fname='data/inflammation-01.csv', delimiter=',')
# Get a sample of the data
# All decimal numbers 0.0 = 0.
```

```python
# Data was loaded but not stored: we need to assign the array to a variable
```

```python
# Star with simple collection of data: single value
weight_kg = 55
```

```python
# print statement (import was also a statement)
print weight_kg
```

```python
# We can print several things (string and numbers) separated by commas, and evaluate arithmetics
print "Weight in pounds:", 2.2 * weight_kg
```

```python
# Assign a new value to the variable
weight_kg = 57.5
print "The weight is now:", weight_kg
```

```python
# Assigning the value to VAR1 does not change the value to VAR2
weight_lb = 2.2 * weight_kg
print 'weight in kilograms:', weight_kg, 'and in pounds:', weight_lb
```

```python
# if I change the value of weight_kg...
weight_kg = 100
print 'weight in kilograms:', weight_kg, 'and in pounds:', weight_lb
```
